# **Preprocessing**

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# Import pandas and read the charity_data.csv from the provided cloud URL.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN','NAME'])
```

```python
# Determine the number of unique values in each column.
unique_count = application_df.nunique()
unique_count
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_counts = application_df['APPLICATION_TYPE'].value_counts()
application_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
type_threshold = 500
application_types_to_replace = list(application_counts[application_counts < type_threshold].index)

# Replace in dataframe
application_df['APPLICATION_TYPE'] =  application_df['APPLICATION_TYPE'].replace(application_types_to_replace, 'Other')

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_count = application_df['CLASSIFICATION'].value_counts()
classification_count
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classificaton_count= classification_count.loc[classification_count > 1]
```

```python
classification_threshold = 1000
classification_to_replace = list(classification_count[classification_count < classification_threshold].index)
application_df['CLASSIFICATION'] =  application_df['CLASSIFICATION'].replace(classification_to_replace, 'Other')
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df, columns=['APPLICATION_TYPE', 'CLASSIFICATION','AFFILIATION','USE_CASE','ORGANIZATION','SPECIAL_CONSIDERATIONS', 'INCOME_AMT'], dtype=int)
application_df
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL']
X = application_df.drop(columns='IS_SUCCESSFUL')

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# **Compile, Train and Evaluate the Model**

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

input_features = len(X_train_scaled[0])
first_layer_nodes_number = 80
second_layer_node_number = 50
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units= first_layer_nodes_number, activation='relu', input_dim=(input_features)))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units= second_layer_node_number, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
import os
os.makedirs("weights", exist_ok=True)
# Custom callback to save weights every 5 epochs
class CustomCheckpoint(tf.keras.callbacks.Callback):
    def __init__(self, save_freq=5):
        self.save_freq = save_freq

    def on_epoch_end(self, epoch, logs=None):
        if (epoch + 1) % self.save_freq == 0:
            filepath = f"weights/weights_epoch_{epoch + 1:02d}.weights.h5"
            self.model.save_weights(filepath)
            print(f"\nSaved weights at epoch {epoch + 1} to {filepath}")
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, callbacks=[CustomCheckpoint(save_freq=5)])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print("Initial Model")
print("------------------------------------------------")
print(f"""Layer1:{first_layer_nodes_number} neurons, relu activation function,
Layer2:{second_layer_node_number} neurons, relu activation function ,
Output Layer: 1 neuron, sigmoid activation function""")
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity.h5", save_format='h5')
```
